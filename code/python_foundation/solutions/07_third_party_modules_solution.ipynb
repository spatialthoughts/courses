```python
from geopy import distance

bangalore = (12.9716, 77.5946)
mysore = (12.2958, 76.6394)

straight_line_distance = distance.great_circle(bangalore, mysore)
ellipsoid_distance = distance.geodesic(bangalore, mysore, ellipsoid='WGS-84')

print(straight_line_distance, ellipsoid_distance)
```

Prompt for Claude AI

I want to use geopy to calculate the distance for the top 5 longest flights.

1. Singapore (SIN) – New York (JFK)
2. Singapore (SIN) – Newark (EWR)
3. Doha (DOH) – Auckland (AKL)
4. Perth (PER) – London Heathrow (LHR)
5. Melbourne (MEL) – Dallas/Fort Worth (DFW)

```python
from geopy import distance

airports = {
    'SIN': (1.3644,  103.9915),   # Singapore Changi
    'JFK': (40.6413, -73.7781),   # New York JFK
    'EWR': (40.6895, -74.1745),   # Newark
    'DOH': (25.2731,  51.6081),   # Doha Hamad
    'AKL': (-37.0082, 174.7917),  # Auckland
    'PER': (-31.9403, 115.9669),  # Perth
    'LHR': (51.4706,  -0.4619),   # London Heathrow
    'MEL': (-37.6690, 144.8410),  # Melbourne
    'DFW': (32.8998, -97.0403),   # Dallas/Fort Worth
}

flights = [
    ('SIN', 'JFK', 'Singapore → New York JFK'),
    ('SIN', 'EWR', 'Singapore → Newark'),
    ('DOH', 'AKL', 'Doha → Auckland'),
    ('PER', 'LHR', 'Perth → London Heathrow'),
    ('MEL', 'DFW', 'Melbourne → Dallas/Fort Worth'),
]

print(f"{'Rank':<5} {'Route':<35} {'Great Circle':>15} {'Geodesic (WGS-84)':>20}")
print("-" * 78)

for i, (orig, dest, label) in enumerate(flights, 1):
    gc  = distance.great_circle(airports[orig], airports[dest])
    geo = distance.geodesic(airports[orig], airports[dest], ellipsoid='WGS-84')
    print(f"{i:<5} {label:<35} {gc.km:>12.1f} km  {geo.km:>16.1f} km")
```
