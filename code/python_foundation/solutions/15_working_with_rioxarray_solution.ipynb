```python
import rioxarray as rxr
import os
import numpy as np

output_filename = 'merged.tif'
output_dir = '../output'
output_path = os.path.join(output_dir, output_filename)

# Read the output file, select the first band, get it as a NumPy array and find the maximum value

# Read the output file, select the first band, get it as a NumPy array and find the maximum value
merged_da = rxr.open_rasterio(output_path)
band = merged_da.sel(band=1)
values = band.values
max_value = np.max(values)
print(max_value)
```

If we wanted to find out the coordinates of Mt. Everest, we can build on this solution.

```python
max_da = merged_da.where(merged_da==merged_da.max(), drop=True).squeeze()
max_x = max_da.x.values
max_y = max_da.y.values
max_elev = int(max_da.values)
print(max_x, max_y, max_elev)
```
