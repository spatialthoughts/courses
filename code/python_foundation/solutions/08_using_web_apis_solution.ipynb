Write Python code to get temperature  forecast at Ahmedabad, India for next 7 days using Open-Meteo https://open-meteo.com/ Weather Forecast API



```python
import urllib.request, json

url = (
    "https://api.open-meteo.com/v1/forecast"
    "?latitude=23.0225&longitude=72.5714"
    "&daily=temperature_2m_max,temperature_2m_min,weathercode"
    "&timezone=Asia%2FKolkata&forecast_days=7"
)

with urllib.request.urlopen(url) as r:
    data = json.loads(r.read())

daily = data["daily"]
print(f"{'Date':<12} {'Max':>6} {'Min':>6}")
print("-" * 26)
for date, tmax, tmin in zip(
    daily["time"],
    daily["temperature_2m_max"],
    daily["temperature_2m_min"],
):
    print(f"{date:<12} {tmax:>5.1f}° {tmin:>5.1f}°")
```
