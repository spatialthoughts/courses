You have the highest elevation in the merged raster. Locate the coordinates of this elevation - which will be at Mt. Everest. Write Python code to read the `merged.tif` file, extract the coordinates at which this elevation occurs and construct a URL to open these coordinates on Google Maps. 

```python
import rioxarray as rxr
import numpy as np
import os

output_filename = 'merged.tif'
output_dir = '../output'
output_path = os.path.join(output_dir, output_filename)

da = rxr.open_rasterio(output_path, masked=True).squeeze()

# Find index of max elevation
max_idx = np.nanargmax(da.values)
row, col = np.unravel_index(max_idx, da.shape)

# Get coordinates
x = float(da.x[col])
y = float(da.y[row])
max_elev = float(da.values[row, col])

print(f"Max elevation: {max_elev:.1f}m")
print(f"Longitude: {x:.6f}, Latitude: {y:.6f}")
print(f"Google Maps URL: https://www.google.com/maps?q={y:.6f},{x:.6f}")
```
